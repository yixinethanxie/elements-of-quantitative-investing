# Proof for Min-Variance Portfolios

**Theorem**:
$$\min w^T\hat{\Omega}_rw$$
$$\text{s.t. } \mu^Tw=1$$

The solution is $w(\hat{\Omega}_r)$. The realized vol of portfolio is greater than that of the portfolio using the empirical cov matrix. Equal when $\propto$ each other

**Proof**:

$$\hat{w} = w(\hat{\Omega}_r) = (\mu^T\hat{\Omega}_r^{-1}\mu)^{-1}\hat{\Omega}_r^{-1}\mu$$
$$w = w(\Omega_r) = (\mu^T\Omega_r^{-1}\mu)^{-1}\Omega_r^{-1}\mu$$

The variance ratio is:
$$\frac{var(\hat{w}, \Omega_r)}{var(w, \Omega_r)} = 
    \frac{\mu^T\Omega_r^{-1}\mu}{\mu^T\hat{\Omega}_r^{-1}\mu}
    \frac{\mu^T\hat{\Omega}_r^{-1}\Omega_r\hat{\Omega}_r^{-1}\mu}{\mu^T\hat{\Omega}_r^{-1}\mu}
$$

If $\hat{\Omega}_r \propto \Omega_r$
$$\frac{var(\hat{w}, \Omega_r)}{var(w, \Omega_r)} = 
    \frac{\mu^T\Omega_r^{-1}\mu}{\mu^T\Omega_r^{-1}\mu}
    \frac{\mu^T\Omega_r^{-1}\Omega_r\Omega_r^{-1}\mu}{\mu^T\Omega_r^{-1}\mu} = 
    \frac{\mu^T\Omega_r^{-1}\mu}{\mu^T\Omega_r^{-1}\mu}
    \frac{\mu^T\Omega_r^{-1}\mu}{\mu^T\Omega_r^{-1}\mu} = 1
$$

Set 
$$\Omega_r = USU^T$$
$$\hat{\Omega}_r = \hat{U}\hat{S}\hat{U}^T$$
$$x = \hat{S}^{-1/2}\hat{U}^T\mu$$
$$H = \hat{S}^{1/2}\hat{U}^TUS^{-1}U^T\hat{U}\hat{S}^{1/2}$$

Then
$$\frac{var(\hat{w}, \Omega_r)}{var(w, \Omega_r)} = 
    \frac{x^THx}{||x||^2}
    \frac{x^TH^{-1}x}{||x||^2}
$$

Set

$$H=VDV^T$$
$$y=V^Tx$$
$$||y||^2 = y^Ty = x^TVV^Tx = x^Tx = ||x||^2$$
$$\frac{var(\hat{w}, \Omega_r)}{var(w, \Omega_r)} = 
    \frac{y^TDy}{||y||^2}
    \frac{y^TD^{-1}y}{||y||^2}
    = 
    \left(\sum_i\frac{y_i^2}{\sum_j y_j^2}d_i\right)
    \left(\sum_i\frac{y_i^2}{\sum_j y_j^2}d_i^{-1}\right)
$$

Let $\xi$ be a random variable that equals $d_i$ with probability $p_i = y_i^2/\sum_j y^2_j$, then
$$\frac{var(\hat{w}, \Omega_r)}{var(w, \Omega_r)} = 
E(\xi)E(1/\xi)$$

$\xi > 0$, to see why
$$H = \hat{S}^{1/2}\hat{U}^TUS^{-1}U^T\hat{U}\hat{S}^{1/2}
    = (\hat{S}^{1/2}\hat{U}^TUS^{-1/2})(\hat{S}^{1/2}\hat{U}^TUS^{-1})^T
$$
H is a Gram matrix, then $H \succeq 0$. H is invertible, so it's full rank and doesn't have zero eigenvalues and $H \succ 0$. Therefore, all eigenvalues $d_i > 0$

Let $f(\xi) = 1/\xi$, where $\xi > 0$. $f(x)$ is a convex function, $f'(\xi) = ln(\xi) > 0$ for $\xi>0$

By Jensen's inequality, $E(f(\xi)) = E(1/\xi) \geq 1/E(\xi)$

Therefore 

$$\frac{var(\hat{w}, \Omega_r)}{var(w, \Omega_r)} = 
E(\xi)E(1/\xi) \geq E(\xi)(1/E(\xi)) = 1$$
